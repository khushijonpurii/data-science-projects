```python
#create dataframe
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/weatherAUS.csv')
df
```

```python
#preprocessing
```

```python
#check for null values
df.isnull().sum()
```

```python
#handle missing values
df=df.dropna(thresh=len(df) * 0.7, axis=1)      #excessive null val col drop
df.fillna(df.mean(),inplace=True)               #fill missing num val with mean
df.fillna(df.mode().iloc[0],inplace=True)       #fill missing categorical val with mode
```

```python
#encoding
from sklearn.preprocessing import LabelEncoder
categorical_columns = df.select_dtypes(include=["object"]).columns
label_encoder= LabelEncoder()
df[categorical_columns]=df[categorical_columns].apply(label_encoder.fit_transform)
```

```python
#split the data
from sklearn.model_selection import train_test_split
x=df.drop("RainTomorrow", axis=1)
y=df["RainTomorrow"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
#logistic regression
```

```python
#apply classifier
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
```

```python
#fit the model
lr_model.fit(x_train_scaled, y_train)
```

```python
#predict the output
lr_y_pred = lr_model.predict(x_test_scaled)
lr_y_pred
```

```python
#accuracy
from sklearn.metrics import accuracy_score
lr_accuracy = accuracy_score(y_test, lr_y_pred)*100
lr_accuracy
```

```python
#knn
```

```python
#apply classifier
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier()
```

```python
#fit the model
knn_model.fit(x_train_scaled, y_train)
```

```python
#predict the output
knn_y_pred = knn_model.predict(x_test_scaled)
knn_y_pred
```

```python
#accuracy
from sklearn.metrics import accuracy_score
knn_accuracy = accuracy_score(y_test, knn_y_pred)*100
knn_accuracy
```

```python
#svm
```

```python
#apply classifier
from sklearn.svm import SVC
svm_model= SVC()
```

```python
#fit the model
svm_model.fit(x_train_scaled, y_train)
```

```python
#predict the output
svm_y_pred = svm_model.predict(x_test_scaled)
svm_y_pred
```

```python
#accuracy
from sklearn.metrics import accuracy_score
svm_accuracy = accuracy_score(y_test, svm_y_pred)*100
svm_accuracy
```

```python
#apply classifier
from sklearn.naive_bayes import GaussianNB
nb_model= GaussianNB()
```

```python
#fit the model
nb_model.fit(x_train_scaled, y_train)
```

```python
#predict the output
nb_y_pred = nb_model.predict(x_test_scaled)
nb_y_pred
```

```python
#accuracy
from sklearn.metrics import accuracy_score
nb_accuracy = accuracy_score(y_test, nb_y_pred)*100
nb_accuracy
```

```python
#comparing accuarcies
```

```python
acc = {}
acc["Logistic Regression"] = lr_accuracy
acc["KNN"] = knn_accuracy
acc["SVM"] = svm_accuracy
acc["Decision Tree"] = dt_accuracy
acc["Random Forest"] = rf_accuracy
```

```python
sorted_acc= sorted(acc.items(), key=lambda x: x[1], reverse=True)
```

```python
for name, accuracy in sorted_acc:
    print(f"{name}: {accuracy}")
```
