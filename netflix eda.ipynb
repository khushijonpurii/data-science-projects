```python
#for analysis3
import numpy as np
import pandas as pd
```

```python
#for visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create a dataframe
df=pd.read_csv('/content/NetflixOriginals.csv',encoding='ISO-8859-1')
df
```

```python
#1.no. of genres
genre= df.Genre.unique()
genre
```

```python
len(genre)
```

```python
#2.How many languages?
lang= pd.unique(df['Language'])
lang
```

```python
len(lang)
```

```python
#3.no. of shows in each genre
df.groupby(['Genre']).size()
```

```python
#4.visualisation for no. of shows in each genre
gvis= df['Genre'].value_counts().head(15).plot(kind='barh',color=['lightblue','pink'],title='NO. OF SHOWS IN EACH GENRE')
display(gvis)
```

```python
#5.sorting data acc to genre
s=df.sort_values('Genre')
s[['Title','Genre']]
```

```python
#6.no. of shows in each language
df.groupby(['Language']).size()
```

```python
#7.visualisation for no. of shows in each language
lvis= df['Language'].value_counts().head(15).plot(kind='barh',color=['lightblue','pink'],title='NO. OF SHOWS IN EACH LANGUAGE')
display(lvis)
```

```python
#8.sorting data acc to language
s=df.sort_values('Language')
s[['Title','Language']]
```

```python
#9.top 10 lengthy shows or movies
l=df.sort_values('Runtime',ascending=False)
lengthy=l[['Title','Runtime']][:10]
lengthy
```

```python
#10.visualisation of lengthy shows
import plotly.express as px
fig_hist=px.bar(lengthy,x='Title',y='Runtime',title='top 10 lenghty movies')
fig_hist.show()
```

```python
#11.average runtime of shows
df['Runtime'].mean()
```

```python
#12.top 10 most rated shows
r=df.sort_values('IMDB Score',ascending=False)
r[['Title','IMDB Score']][:10]
```

```python
#13.visualisation of top 10 most rated shows
import plotly.express as px
rating = r[['Title','IMDB Score']][:10]
fig_hist=px.bar(rating,x='Title',y='IMDB Score',title='top 10 most rated shows')
fig_hist.show()
```

```python
#14.shows with imdb score greater than 8.5
dhigh=df[df['IMDB Score'].values>=8.5]
dhigh
```

```python
#15.shows with imdb score less than 3
dlow=df[df['IMDB Score'].values<=3.0]
dlow
```

```python
#16.visualisation of imdb score count
sns.countplot(x=df['IMDB Score'])
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title('IMDB SCORE COUNT')
```
