```python
#Q5: Top 5 run scorers
top_run_scorers = batting_df.sort_values(by='runs', ascending=False).head(5)
top_run_scorers
```

```python
#Q6: Plot a graph for the top 5 run scorers in the series
plt.bar(top_run_scorers['Player'], top_run_scorers['runs'],color='violet')
plt.xlabel('player')
plt.ylabel('runs')
plt.title('TOP 5 RUN SCORERS')
plt.show()
```

```python
#Q7: Top 5 wicket takers
top_wicket_takers = bowling_df.sort_values(by='wickets', ascending=False).head(5)
top_wicket_takers
```

```python
#Q8: Plot the top 5 wicket takers of the series (Bar Graph)
plt.bar(top_wicket_takers['Player'],top_wicket_takers['wickets'],color='khaki')
plt.xlabel('Player')
plt.ylabel('wickets')
plt.title('TOP 5 WICKET TAKERS')
plt.show()
```

```python
#Q9: Best All-round performers
all_rounders_perf = combined_df[(combined_df['bat_avg'] > 20) & (combined_df['bw_avg'] >= 20) & (combined_df['bw_avg'] <= 30)]
all_rounders_perf
```

```python
#Q10: Plot the best allrounder graph (Bar Graph)
plt.bar(all_rounders_perf['Player'], all_rounders_perf['runs_scored'],color='pink')
plt.xlabel('Player')
plt.ylabel('run scored')
plt.title('BEST ALL ROUNDER PERFORMANCE')
plt.show()
```

```python
#Q11: Who has the best batting average?
best_batting_avg = batting_df.sort_values(by='best', ascending=False).iloc[0]['Player']
best_batting_avg
```

```python
#Q12: Who has the best bowling average?
best_bowling_avg = bowling_df.sort_values(by='best').iloc[0]['Player']
best_bowling_avg
```

```python
#Q13: Who has the highest strike rate?
highest_strike_rate = batting_df.sort_values(by='strike_rate', ascending=False).iloc[0]['Player']
highest_strike_rate
```

```python
#Q14: How many people have scored centuries? List them
century_scorers = batting_df[batting_df['hundreds'] > 0]
n_century_scorers = century_scorers.shape[0]
century_scorers_list = century_scorers['Player'].tolist()
```

```python
#Q15: How many people have got 5-wicket hauls?
n_five_wicket_hauls = bowling_df[bowling_df['fives'] > 0].shape[0]
n_five_wicket_hauls
```

```python
#Q16: Who has got the highest wickets in an inning?
highest_wickets_in_inning = bowling_df.sort_values(by='innings', ascending=False).iloc[0]['Player']
highest_wickets_in_inning
```

```python
#Q17: Who has got the best figure? (More wickets and fewer runs given)
bowling_df['economy'] = bowling_df['runs'] / bowling_df['wickets']
best_figure_player = bowling_df.sort_values(by=['wickets', 'economy'], ascending=[False, True]).iloc[0]['Player']
best_figure_player
```

```python
#Q18: Percentage of runs and wickets for teams
england_runs_percentage = (batting_df.loc[:17, 'runs'].sum() / batting_df['runs'].sum()) * 100
india_runs_percentage = (batting_df.loc[17:, 'runs'].sum() / batting_df['runs'].sum()) * 100

england_wickets_percentage = (bowling_df.loc[:17, 'wickets'].sum() / bowling_df['wickets'].sum()) * 100
india_wickets_percentage = (bowling_df.loc[17:, 'wickets'].sum() / bowling_df['wickets'].sum()) * 100
```

```python
england_wickets_percentage
```

```python
india_wickets_percentage
```

```python
#Q19: Plot a pie chart for displaying the England run percentage
labels = ['England', 'Rest of the Teams']
sizes = [england_runs_percentage, 100 - england_runs_percentage]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%',colors=['hotpink','khaki'])
plt.title('England Run Percentage')
plt.show()
```

```python
#Q20: Plot a pie chart for displaying the India run percentage
labels = ['India', 'Rest of the Teams']
sizes = [india_runs_percentage, 100 - india_runs_percentage]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%',colors=['pink','yellow'])
plt.title('India Run Percentage')
plt.show()
```
