```python
#col of intersest
selected_columns = ['iyear', 'imonth', 'iday', 'country_txt', 'region', 'INT_LOG', 'INT_IDEO']
```

```python
#counts and unique values
for column in selected_columns:
    print("Column:", column)
    print(df[column].value_counts())
    print()
```

```python
# Calc for selected columns
print("Descriptive statistics for selected columns:")
print(df[selected_columns].describe())
```

```python
#visualisation
import matplotlib.pyplot as plt
```

```python
# Plot a bar chart of terrorism incidents by year
attacks_by_year = df['iyear'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
plt.bar(attacks_by_year.index, attacks_by_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.title('Terrorist Attacks by Year')
plt.show()
```

```python
# Plot a pie chart of terrorism incidents by region
attacks_by_region = df['region'].value_counts()
plt.figure(figsize=(12,12))
plt.pie(attacks_by_region.values, labels=attacks_by_region.index, autopct='%1.1f%%')
plt.title('Terrorist Attacks by Region')
plt.show()
```
